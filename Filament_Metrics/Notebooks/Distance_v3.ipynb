```python
def galfildist3(galaxies, filament_points): ##USE THIS ONE W OPTIMIZATION 
     '''Finds the closest distances between an array of galaxies
        and array of filament points. "filpoints" array must be a
         simple 3d array of all coordinate points of a filament
          (in other words - not an array of arrays, which separates half filaments) 
          MUST GIVE A SPECIFIC TYPE OF ARRAY WHERE EVERY ELEMENT IS AN ARRAY OF SEGMENTS OF FILAMENT'''
     shortest_distances = np.zeros(len(galaxies))
     for h in range(len(galaxies)):
            distances = np.zeros(len(filament_points))
            for i in range(len(filament_points)):
                  A = filament_points[i][:-4] ; B = filament_points[i][3:-1] ; AB = B - A ; AG = galaxies[h] - A
                  ## OPTIMIZATION STEP 
                  if np.any(AG > 10) or np.any(AG < -10):
                        distances[i] = 100
                        continue
                  # get norm -- in old code had to check if AB == 0 
                  u = A / np.linalg.norm(A)
                  #if statements to check separate cases
                  if np.dot(AG, u) < 0:
                        distances[i] = np.linalg.norm(galaxies[h] - A)
                  elif np.dot(AG, u) > np.linalg.norm(AB):
                        distances[i] = np.linalg.norm(galaxies[h] - B)
                  elif np.dot(AG, u) < np.linalg.norm(AB):
                        distances[i] = np.linalg.norm((np.cross((A-B), (A-galaxies[h])))) / (np.linalg.norm(galaxies[h]-B))
            shortest_distances[h] = np.min(distances)
     return shortest_distances
```
